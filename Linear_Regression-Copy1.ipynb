```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(10)
X= 2 * np.random.randn(500,1)
y = 4 + 3 * X + np.random.randn(500,1)
plt.scatter(X,y)
plt.show()
```

```python
X_b = np.c_[np.ones((500,1)),X]
print(X_b.shape, y.shape)
```

```python
learning = 0.1
iterations = 500
m= 500
theta = np.random.randn(2,1)
theta
```

```python
for iterations in range(iterations):
    gradient = 2/m * X_b.T.dot(X_b.dot(theta)-y)
    theta = theta-learning *gradient
    print(theta)
```

```python
intercept,slope=theta
```

```python
print("Intercept : ", intercept)
print("Slope : ", slope)
```

```python
for i in range(len(X)):
    y_p = slope*X[i] + intercept
    print(y_p)
```

```python
import matplotlib.pyplot as plt

y_p = slope*X + intercept
plt.plot(X,y_p,color = 'blue')
plt.scatter(X,y, color = 'red')
plt.show()
```
